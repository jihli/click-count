Section 3: Merge "site_search_result" with "after search product view" (result is site_search_result_after)

```python
import pandas as pd

# Load the CSV files
site_search_result_df = pd.read_csv('site_search_result.csv')
after_search_df = pd.read_csv('Event - After Search Product Click.csv')

# Convert event_timestamp to datetime format for both dataframes
site_search_result_df['event_timestamp'] = pd.to_datetime(site_search_result_df['event_timestamp'], format='%Y/%m/%d/%H/%M/%S')
after_search_df['event_timestamp'] = pd.to_datetime(after_search_df['event_timestamp'], format='%Y/%m/%d/%H/%M/%S')

# Identifying and merging matched rows
matched_rows = []
for after_idx, after_row in after_search_df.iterrows():
    condition = (
        (site_search_result_df['customer_id'] == after_row['customer_id']) &
        (site_search_result_df['ga_session_id'] == after_row['ga_session_id']) &
        (site_search_result_df['event_timestamp'] < after_row['event_timestamp'])
    )
    filtered_site_search = site_search_result_df[condition].copy()
    if not filtered_site_search.empty:
        closest_row = filtered_site_search.iloc[-1]  # The most recent one before after_search event
        closest_row_index = closest_row.name
        for col in after_search_df.columns:
            closest_row[f"after_{col}"] = after_row[col]
        matched_rows.append(closest_row)

matched_df = pd.DataFrame(matched_rows)

# Include all site_search_result rows
# First, set a flag for matched rows to exclude them from being added again
site_search_result_df['matched'] = site_search_result_df.index.isin(matched_df.index)

# For rows in site_search_result_df that weren't matched, add them with placeholders for after_search columns
unmatched_df = site_search_result_df[~site_search_result_df['matched']].copy()
for col in after_search_df.columns:
    unmatched_df[f"after_{col}"] = None
unmatched_df.drop(columns=['matched'], inplace=True)

# Combine matched and unmatched rows
final_df = pd.concat([matched_df, unmatched_df], ignore_index=True)
```

```python
site_search_result_after = final_df
site_search_result_after.head(5)
```

Section 4: Merge "site_search_result_after" with "Event - Product View" (result is site_search_result_after_view)

```python
import pandas as pd

product_view[['product_name', 'product_sku']] = product_view['event_value'].str.split('\|\|\|', expand=True)
unique_products = product_view[['product_name', 'product_sku']].drop_duplicates()

# 存储到新的变量
unique_product_details = unique_products

print(unique_product_details.head(5))
```

```python
import pandas as pd

site_search_result_after['after_event_value'] = site_search_result_after['after_event_value'].str.strip()

# 去除unique_product_details中product_name列的首尾空格
unique_product_details['product_name'] = unique_product_details['product_name'].str.strip()

# 现在执行左连接
site_search_result_after_view = pd.merge(site_search_result_after, unique_product_details, how='left', left_on='after_event_value', right_on='product_name')

print(site_search_result_after_view.head())
```
